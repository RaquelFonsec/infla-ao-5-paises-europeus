```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de pares de países
pares_paises = [
    ('Alemanha', 'Franca'),
    ('Alemanha', 'Belgica'),
    ('Alemanha', 'Holanda'),
    ('Alemanha', 'Luxemburgo'),
    ('Franca', 'Belgica'),
    ('Franca', 'Holanda'),
    ('Franca', 'Luxemburgo'),
    ('Belgica', 'Holanda'),
    ('Belgica', 'Luxemburgo'),
    ('Holanda', 'Luxemburgo')
]

# Armazenar resultados
resultados = []

# Realizar o teste t para cada par de países
for pais1, pais2 in pares_paises:
    # Filtrando os dados de inflação mensal dos dois países
    dados_pais1 = globals()[f"df_{pais1.lower()}_2021_2023"]['Inflação Mensal (%)']
    dados_pais2 = globals()[f"df_{pais2.lower()}_2021_2023"]['Inflação Mensal (%)']

    # Realizando o teste t de Student para as duas amostras independentes
    t_stat, p_value = stats.ttest_ind(dados_pais1, dados_pais2)

    # Definir o nível de significância (0.05)
    alpha = 0.05

    # Armazenar o resultado
    resultados.append({
        'Países': f'{pais1} vs {pais2}',
        'P-valor': p_value,
        'Significativo': 'Sim' if p_value < alpha else 'Não'
    })

# Criar DataFrame para resultados
df_resultados = pd.DataFrame(resultados)

# Plotando o gráfico com 'hue' para diferenciar as barras
plt.figure(figsize=(10, 6))
sns.barplot(x='P-valor', y='Países', data=df_resultados, palette='coolwarm', hue='Significativo', dodge=False)

# Adicionando título e rótulos
plt.title('Resultados dos Testes T de Inflação Mensal entre os Países', fontsize=14)
plt.xlabel('P-valor', fontsize=12)
plt.ylabel('Comparação de Países', fontsize=12)

# Exibir p-valor nas barras
for i, p_value in enumerate(df_resultados['P-valor']):
    plt.text(p_value + 0.001, i, f'{p_value:.4f}', va='center', fontsize=10)

# Mostrar o gráfico
plt.show()
```

Conclusões Gerais:

Diferença Estatisticamente Significativa: A única comparação que resultou em uma diferença estatisticamente significativa entre as médias de inflação mensal foi entre França e Bélgica (p-valor: 0.0125), indicando que as médias de inflação mensal desses dois países são significativamente diferentes.


Sem Diferença Estatística Significativa: Para todos os outros pares de países analisados, os p-valores foram superiores ao nível de significância de 0.05, o que significa que não há diferença estatística significativa entre suas médias de inflação mensal.
